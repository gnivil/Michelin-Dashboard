```python
#Run for loop to scrape data for three star restaurants
Restaurants = []

for link in three_star_list:
    browser.visit(link)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    # print(link)
    r_name = (soup.find_all('h2', class_="restaurant-details__heading--title")[0].text)
    price = soup.find('li', class_="restaurant-details__heading-price").contents[0]
    price = price.strip()
    price_list = price.split()
    price_list = price_list[0:price_list.index('•')]
    price_range = ' '.join(price_list)   
    try:
        external_url = (soup.find_all('span', class_="flex-fill")[1].text)
    except Exception as e:
        external_url = 'N/A'
        print(e)
    if not (external_url.startswith('www') or external_url.startswith('http')):
        external_url = 'N/A'
    print(external_url)
    Restaurants.append({'Restaurant Name': r_name,
                        'Price Range': price_range,
                        'Restauarant Website' : external_url})
```

```python
import re
#for loop to scrape lat and long for three star restaurants
three_star_lat_dict = []
for link in three_star_list:
    browser.visit(link)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    r_name = soup.find_all('h2', class_="restaurant-details__heading--title")[0].text
    lat_long = soup.find('div', class_='google-map__static')
    lat_long_url = lat_long.find('iframe')['src']
    try: 
        if ",-" or "=-" in lat_long_url:
            lat_long_regex = re.findall(r'-?\d+?\d*', lat_long_url)
            lat = lat_long_regex[4]
            long = lat_long_regex[6]
        else:
            lat_long_regex = re.findall(r'\b\d+\b',lat_long_url)
            lat = lat_long_regex[0]
            long = lat_long_regex[2]
    except Exception as e:
        lat = 'N/A'
        long = 'N/A'
        print(e,r_name)
    three_star_lat_dict.append({'Restaurant Name': r_name,
                        'Latitude': lat,
                        'Longitude' : long})
```
